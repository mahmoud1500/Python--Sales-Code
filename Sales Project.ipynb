```python
print("Create Sales Program")

a = {}

with open(r"C:\Mahmoud\bellevuecollege\spring2022\PROG108\pythonscripts\monthly_sales.txt", "r", encoding="utf-8") as f:
    for line in f:
        month, value = line.split()
        a[month] = int(value)

print("press 1 for the total annual sales")
print("press 2 for the average annual sales")
print("press 3 for the month with highest sales")

code = int(input("Enter number: "))

if code == 1:
    total = sum(a.values())
    print("Total annual sales:", total)

elif code == 2:
    average = sum(a.values()) / len(a)
    print("Average annual sales:", average)

elif code == 3:
    highest_month = max(a, key=a.get)
    print("Month with highest sales:", highest_month, "=", a[highest_month])

else:
    print("Invalid choice")
```
